## 折线图pyplot.plot(), 散点图pyplot.scatter()

```python
import matplotlib as mpl 
import matplotlib.pyplot as plt
#matplotlib 可以绘制非常绚丽的图，但是我们这里只需要使用最贱的pyplot
```

```python
import numpy as np
x = np.linspace(0, 10, 101)
```

```python
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y) 
```

```python
plt.show()
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red')
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle = '--') # ':', '-.', '--','-'
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle = '--')
plt.xlim(-5, 15)#x-axis limit
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle = '--')
plt.xlim(-5, 15) #x-axis limit
plt.ylim(0, 1.5)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle = '--')
plt.axis([-1, 11, -2 ,2]) #前两个数字是x轴范围：[-1,11], 后两个数字是y轴范围：[-2,2]
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle = '--')
plt.xlabel("x axis") #坐标的名称
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x, siny, label = "sin(x)")
plt.plot(x, cosy, color = 'red', linestyle = '--', label = "cos(x)") #label需要加plt.legend()
plt.xlabel("x axis") #坐标的名称
plt.ylabel("y value")
plt.title("welcome")
plt.legend()
plt.show()
```

### Scatter plot

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color = 'red')
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)

plt.scatter(x, y)
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)

plt.scatter(x, y, alpha = 0.1) #设置透明度为0.1
plt.show()
```
