```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = np.random.uniform(-3., 3., size = 100)
X = x.reshape(-1,1)
y = 0.5 * X + 3 + np.random.normal( size = (100,1))
```

```python
plt.scatter(X, y)
plt.show()
```

### Polynomial Regression

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

def PolynomialRegression(degree):
    return Pipeline([
    ("my_poly", PolynomialFeatures(degree) ), #tuple1： （"自己命名"， 类名）
    ("my_std_scaler", StandardScaler() ),
    ("my_lin_reg", LinearRegression() )
    ])
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(666)
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.metrics import mean_squared_error 

poly_reg = PolynomialRegression(degree=20)
poly_reg.fit(X_train,y_train)

y_poly_predict = poly_reg.predict(X_test)
print("MSE:", mean_squared_error(y_test, y_poly_predict))
```

```python
X_plot = np.linspace(-3, 3, 100).reshape(100,1)
y_plot = poly_reg.predict(X_plot)

plt.scatter(x,y)
plt.plot(X_plot[:,0], y_plot, color = 'r')
plt.axis([-3,3,0,6])
plt.show()
```

```python
def plot_model(model):
    X_plot = np.linspace(-3,3,100).reshape(100,1)
    y_plot = model.predict(X_plot)

    plt.scatter(x,y)
    plt.plot(X_plot[:,0], y_plot, color = 'r')
    plt.axis([-3,3,0,6])
    plt.show()
    
plot_model(poly_reg)
```

### Using LASSO
1. lasso 需要的alpha值 比ridge的alpha值大
2. 因为ridge是theta的平方，所以需要缩小的大一些，需要的alpha也相应的小
3. 或者说，ridge是一个大胖子，lasso是一个正常人。alpha小说明食物少，alpha大说明食物多。
4. 对于lasso来说alpha=1会饿的死去，但是对于ridge来说alpha=1是可以忍受的因为它体内很多能量

```python
from sklearn.linear_model import Lasso

def RidgeRegression(degree, alpha):
    return Pipeline([
        ("my_poly", PolynomialFeatures(degree = degree) ), #tuple1： （"自己命名"， 类名）
        ("my_std_scaler", StandardScaler() ),
        #("my_lin_reg", LinearRegression() ),#不能有这一句，需要从StandardScaler()直接到Ridge()
        ("ridge_reg", Lasso(alpha = alpha))
    ])
```

```python
lasso1_reg = RidgeRegression(20, 0.01)
```

```python
lasso1_reg.fit(X_train, y_train)

y1_predict = lasso1_reg.predict(X_test)
mean_squared_error(y_test, y1_predict) #比之前MSE: 167.9401085999025好很多。
```

```python
plot_model(lasso1_reg)
```

```python
lasso2_reg = RidgeRegression(20, 1)
lasso2_reg.fit(X_train, y_train)

y2_predict = lasso2_reg.predict(X_test)
print("MSE:", mean_squared_error(y_test, y2_predict))

plot_model(lasso2_reg)
```

```python
lasso3_reg = RidgeRegression(20, 0.5)
lasso3_reg.fit(X_train, y_train)

y3_predict = lasso3_reg.predict(X_test)
print("MSE:", mean_squared_error(y_test, y3_predict))

plot_model(lasso3_reg) #比之前MSE: 1.1888759304218461 高，说明惩罚过度。
```

```python
ridge4_reg = RidgeRegression(20, 10000000)
ridge4_reg.fit(X_train, y_train)

y4_predict = ridge4_reg.predict(X_test)
print("MSE:", mean_squared_error(y_test, y4_predict))

plot_model(ridge4_reg) #我们的theta1 == 0（注意没有theta2，theta3，说明完全不受X值影响）
```
